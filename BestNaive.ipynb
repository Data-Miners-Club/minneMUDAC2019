```
import os
import random as rn

path = os.path.join("drive", "My Drive", "MinneMUDAC2019", "newFiles")
march = pd.read_csv(os.path.join(path,"march_new.csv"),skiprows = 3)
may = pd.read_csv(os.path.join(path,"may_new.csv"),skiprows = 3)
july = pd.read_csv(os.path.join(path,"july_new.csv"),skiprows = 3)
window = 5
july_close = july["Close"].values[674-window:].tolist()
may_close = may["Close"].values[422-window:].tolist()
march_close = march["Close"].values[422-window:].tolist()
```

```
num_weeks = 16
print(len(july_close))
def exponential(data, num_weeks, filename=""):
  predictions = []
  day_of_week_error = [0, 0, 0, 0, 0]
  weekly_abs_error_sum = []
  if num_weeks ==16:
    data.append(data[-1])
    data.append(data[-1])
    data.append(data[-1])
    data.append(data[-1])
    data.append(data[-1])
  for i in range(num_weeks): # week loop
    prev = data[i*5: i*5 + window]
    abs_error = []
    for j in range(5): # loop through days of the week
      pred_val = np.array(prev).mean()
      predictions.append(pred_val) # append predicted value
      if i != 15:
        diff = pred_val - data[window + i*5 + j] # cal diff
        day_of_week_error[j] = day_of_week_error[j] + diff # sum day pf week error
        abs_error.append(abs(diff)) # save abs error
    weekly_abs_error_sum.append(np.array(abs_error).sum()) # summ weekly abs error

  dic = {"Actual": data[window:], "forecast": predictions}
  results = pd.DataFrame.from_dict(dic)
  results.index.name = 'Period'
  results["error"] = abs(results["Actual"] - results["forecast"])
  if filename != "":
    results.to_csv(filename+".csv")
  return results, day_of_week_error, weekly_abs_error_sum
```

```
julyR, julyDayErr, JulyAbsErr = exponential(july_close, num_weeks, "july_naive_mean")

mayR, mayDayErr, mayAbsErr = exponential(may_close, num_weeks, "may_naive_mean")
marchR, marchDayErr, marchAbsErr = exponential(march_close, num_weeks, "march_naive_mean")
print(np.array(marchAbsErr).sum()/(75))
```
